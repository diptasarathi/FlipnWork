1. Scrape the details of most viewed videos on YouTube from Wikipedia:
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos

```python
driver = webdriver.Edge("msedgedriver.exe")
url="https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)

driver.find_element_by_xpath("//b").click()

time.sleep(3)
records = []
for row in driver.find_elements_by_xpath("(//table[@class='wikitable sortable jquery-tablesorter'])[1]//tr")[1:-1]:
    temp = ''
    for element in row.find_elements_by_xpath(".//td"):        
        temp += element.text + '|'
    records.append(temp.split('|')[:-2])

df = pd.DataFrame(records,columns=['rank','video_name','artist','views(M)','date_uploaded'])
    
arranged_cols = ['rank','video_name','artist','date_uploaded','views(M)']
    
df.video_name = df.video_name.apply(lambda x:x[:-4].strip('"'))
df = df[arranged_cols]
df
    
```

9.Scrape the details most watched tv series of all time from imdb.com.

```python
driver = webdriver.Chrome(r"chromedriver_win32\chromedriver.exe")
url="https://www.imdb.com/list/ls095964455"
driver.get(url)
time.sleep(5)
```

```python
Name=[]
Year_span=[]
Genres=[]
Run_time=[]
Ratings=[]
Votes=[]
#scraping the Name 
mname=driver.find_elements_by_xpath('//h3[@class="lister-item-header"]/a')
for i in mname:
    if i.text is None :
        Name.append("--") 
    else:
        Name.append(i.text)
#scraping the Year_span 
ys=driver.find_elements_by_xpath("//span[@class='lister-item-year text-muted unbold']")
for i in ys:
    if i.text is None :
        Year_span.append("--") 
    else:
        Year_span.append(i.text)
#scraping the Genres 
gnr=driver.find_elements_by_xpath("//span[@class='genre']")
for i in gnr:
    if i.text is None :
        Genres.append("--") 
    else:
        Genres.append(i.text)
#scraping the Run_time 
rt=driver.find_elements_by_xpath("//span[@class='runtime']")
for i in rt[0:101]:
    if i.text is None :
        Run_time.append("--") 
    else:
        Run_time.append(i.text)
#scraping the Ratings 
rate=driver.find_elements_by_xpath("//div[@class='ipl-rating-star small']/span[2]")
for i in rate:
    if i.text is None :
        Ratings.append("--") 
    else:
        Ratings.append(i.text)

#scraping the Votes 
v=driver.find_elements_by_xpath("//div[@class='lister-item-content']/p[4]/span[2]")
for i in v:
    if i.text is None :
        Votes.append("--") 
    else:
        Votes.append(i.text)
        
IMDB_TOP_100_Series=pd.DataFrame([])
IMDB_TOP_100_Series['Name']=Name
IMDB_TOP_100_Series['Year_span']=Year_span
IMDB_TOP_100_Series['Run_time']=Run_time
IMDB_TOP_100_Series['Genres']=Genres
IMDB_TOP_100_Series['Ratings']=Ratings
IMDB_TOP_100_Series['Votes']=Votes
IMDB_TOP_100_Series
```

10.Details of Datasets from UCI machine learning repositories.
 Url = https://archive.ics.uci.edu/

```python
driver = webdriver.Chrome(r"chromedriver_win32\chromedriver.exe")
url="https://archive.ics.uci.edu/"
driver.get(url)
time.sleep(5)
```

```python
search_1 = driver.find_element_by_xpath('/html/body/table[1]/tbody/tr/td[2]/span[2]/a')       # Locating page foe top videos by xpath
search_1.click()
```

```python

Dataset_Name=[]
Data_Type=[]
Task=[]
Attribute_Type=[]
No_of_Instances=[]
No_of_Attribute=[]
Year=[]


#scraping the Dataset_Name 
dname=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b/a")
for i in dname:
    if i.text is None :
        Dataset_Name.append("--") 
    else:
        Dataset_Name.append(i.text)
#scraping the Data_Type 
dtype=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]")
for i in dtype:
    if i.text is None :
        Data_Type.append("--") 
    else:
        Data_Type.append(i.text)

#scraping the Task 
t=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]")
for i in t:
    if i.text is None :
        Task.append("--") 
    else:
        Task.append(i.text)
#scraping the Attribute_Type 
att=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]")
for i in att:
    if i.text is None :
        Attribute_Type.append("--") 
    else:
        Attribute_Type.append(i.text)

#scraping the No_of_Instances 
noi=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]")
for i in noi:
    if i.text is None :
        No_of_Instances.append("--") 
    else:
        No_of_Instances.append(i.text)
#scraping the No_of_Attribute 
noa=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]")
for i in noa:
    if i.text is None :
        No_of_Attribute.append("--") 
    else:
        No_of_Attribute.append(i.text)
#scraping the Year 
y=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]/p")
for i in y:
    if i.text is None :
        Year.append("--") 
    else:
        Year.append(i.text)

UCI=pd.DataFrame([])
UCI['Dataset_Name']=Dataset_Name[:100]
UCI['Data_Type']=Data_Type[:100]
UCI['Task']=Task[:100]
UCI['Attribute_Type']=Attribute_Type[:100]
UCI['No_of_Instances']=No_of_Instances[:100]
UCI['No_of_Attribute']=No_of_Attribute[:100]
UCI['Year']=Year[:100]
UCI
```
