```python
#Q6 : Scrape data of first 100 sunglasses listings on flipkart.com. You have to 
#scrape four attributes:

driver = webdriver.Edge('msedgedriver.exe')
url='https://www.flipkart.com/'
driver.get(url)
time.sleep(5)
login_btn = driver.find_element_by_xpath('//button[@class="_2KpZ6l _2doB4z"]')
login_btn.click()
time.sleep(5)
search_box = driver.find_element_by_class_name('_3704LK')
search_box.send_keys('sunglasses')
time.sleep(2)
search_btn = driver.find_element_by_xpath('//button[@class="L0Z3Pu"]')
search_btn.click()
time.sleep(15)
disc_sg = []
price_sg = []
brand_sg = []
desc_sg = []
while len(disc_sg)<100:
    spec_brand = driver.find_elements_by_xpath('//div[@class="_2WkVRV"]')
    for i in spec_brand :
        brand_sg.append(i.text)
    spec_desc =  driver.find_elements_by_xpath('//a[@class="IRpwTa"]')
    for i in spec_desc :
        desc_sg.append(i.text)
    spec_price = driver.find_elements_by_xpath('//div[@class="_30jeq3"]')
    for i in spec_price :
        price_sg.append(i.text)
    spec_disc = driver.find_elements_by_xpath('//div[@class="_3Ay6Sb"]// span')
    for i in spec_disc :
        disc_sg.append(i.text)
    next_lnk = driver.find_elements_by_class_name("_1LKTO3")
    lnk = []
    for i in next_lnk:
        lnk.append(i)
    lnk[-1].click()
    time.sleep(5)
FLIPKART_SUNGLASS= pd.DataFrame()
FLIPKART_SUNGLASS['Brand'] =brand_sg[:100]
FLIPKART_SUNGLASS['Product Description']=desc_sg[:100]
FLIPKART_SUNGLASS['Price']=price_sg[:100]
FLIPKART_SUNGLASS['Discount %']=disc_sg[:100]
FLIPKART_SUNGLASS
```

```python
# Go to the link - https://www.myntra.com/shoes
#Set Price filter to “Rs. 6649 to Rs. 13099” , Color filter to “Black”, as shown in the below image

driver = webdriver.Edge("msedgedriver.exe")
url='https://www.myntra.com/shoes'
driver.get(url)
time.sleep(15)
#Price_filter = driver.find_element_by_xpath('//* [text()="Rs. 6649 to Rs. 13099"]')
Price_filter = driver.find_element_by_xpath('//* [text()="Rs. 6612 to Rs. 13075"]')
Price_filter.click()
time.sleep(30)
Colour_filter = driver.find_element_by_xpath('//*[text()="Black"]')
Colour_filter.click()
time.sleep(30)
product_brand = driver.find_elements_by_xpath('//h3 [@class="product-brand"]')
Shoe_brand =[]
for i in product_brand :
    Shoe_brand.append(i.text)
product_shrt_desc = driver.find_elements_by_xpath('//h4 [@class="product-product"]')
shrt_desc =[]
for i in product_shrt_desc :
    shrt_desc.append(i.text)
product_price = driver.find_elements_by_xpath('//div [@class="product-price"]')
price =[]
for i in product_price :
    price.append(i.text.replace('Rs. ',' Rs.'))
Myntra_shoes= pd.DataFrame()
Myntra_shoes ['Shoe_brand']=Shoe_brand
Myntra_shoes['shrt_desc']=shrt_desc
Myntra_shoes['price']=price
Myntra_shoes
```
